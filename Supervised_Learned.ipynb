```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn import cluster, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
%matplotlib inline

#Import pytorch:
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torchvision
import torchvision.transforms as transforms
from torch.autograd import Variable
```

```python
from src.spiking_network.network.neuvisys import SpikingNetwork
from src.spiking_network.analysis.spike_train import (
    spike_trains,
    kernel_convolution)
```

```python
kernel_size = 20
sampling_size = 10
```

```python
print("Loading network...")
network_path = "/home/thomas/Networks/simulation/supervised_learning/learned/network_clockwise/"
spinet = SpikingNetwork(network_path, loading=[True, True, False, False])

print("Kernel convolution...")
sts = spike_trains(spinet.spikes[0])
X1 = kernel_convolution(sts, kernel_size, sampling_size)

ori = np.array(spinet.state["learning_data"]["reward"])
idx = np.round(np.linspace(0, len(ori) - 1, X1.shape[0])).astype(int)

Y1 = np.abs(ori[idx])
X1 = X1 / np.max(X1)

print(X1.shape, Y1.shape)

print("Loading network...")
network_path = "/home/thomas/Networks/simulation/supervised_learning/learned/network_counterclockwise/"
spinet = SpikingNetwork(network_path, loading=[True, True, False, False])

print("Kernel convolution...")
sts = spike_trains(spinet.spikes[0])
X2 = kernel_convolution(sts, kernel_size, sampling_size)

ori = np.array(spinet.state["learning_data"]["reward"])
idx = np.round(np.linspace(0, len(ori) - 1, X2.shape[0])).astype(int)

Y2 = np.abs(ori[idx])
X2 = X2 / np.max(X2)

print(X2.shape, Y2.shape)

X = np.concatenate((X1, X2[:X1.shape[0]]))
Y = np.concatenate((Y1, Y2[:Y1.shape[0]]))

print(X.shape, Y.shape)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.layer = nn.Sequential(
#             nn.Linear(36864, 1),
            nn.Linear(5000, 1),
        )
        
    def forward(self, x):
        x = self.layer(x)
        return x

torch.manual_seed(0)
model = Net()
print(model)
```

```python
N_epochs = 1000
batch_size = 100
learning_rate = 0.001

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
loss_func = torch.nn.MSELoss()
```

```python
x = torch.from_numpy(X).float()
y = torch.from_numpy(Y.reshape(-1,1)).float()
print(x.shape, y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape, y_train.shape)

inputs = Variable(x_train[:, :5000])
outputs = Variable(y_train)
inputs_test = Variable(x_test[:, :5000])
outputs_test = Variable(y_test)
```

```python
losses = []
for i in range(N_epochs):
    prediction = model(inputs)
    
    loss = loss_func(prediction, outputs)
    losses.append(loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if i % 50 == 0:
        plt.scatter(y_train.data.numpy(), prediction.data.numpy())
        plt.text(1.1, 0, 'Loss=%.6f' % loss.data.numpy(), fontdict={'size': 10, 'color':  'red'})
    plt.show()
```

```python
# Afficher l'volution de la fonction de cot
fig, axes = plt.subplots(figsize=(8,6))
axes.plot(losses, 'r-', lw=2, label='Loss function')
axes.set_xlabel('N iterations', fontsize=20)
axes.set_ylabel('Loss', fontsize=20)
plt.legend(loc='upper right', fontsize=16)
plt.show()

fig, axes = plt.subplots(figsize=(8,6))
axes.plot(losses[0:50], 'r-', lw=2, label='Loss function')
axes.set_xlabel('N iterations', fontsize=18)
axes.set_ylabel('Loss', fontsize=18)
plt.legend(loc='upper right', fontsize=16)
plt.show()
```
